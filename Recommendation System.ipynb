```python
user_df = df.pivot(index='anime_id', 
                                 columns='genre', 
                                 values='rating')
```

```python
user_df
```

```python
user_df.shape
```

```python
user_df.fillna(0, inplace=True)
```

```python
user_df
```

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python

user_sim = 1 - pairwise_distances(user_df.values,metric='cosine')
```

```python
user_sim
```

```python
np.fill_diagonal(user_sim, 0)
```

```python
user_sim
```

```python

user_sim_df = pd.DataFrame(user_sim)
```

```python
user_sim_df
```

```python

user_sim_df.index = df.anime_id.unique()
user_sim_df.columns = df.anime_id.unique()
```

```python
user_sim_df
```

```python

user_sim_df.idxmax(axis=1)
```

```python
1) can you explain the difference between user-based and item-based collaborative filtering?
              User-based collaborative filtering recommends items based on similarities between users,finding users with similar preferences and suggesting 
what they linked.It struggles with new users and scalability as the user base grows.Item-based collaborative filtering,on the other hand,recommends
items similar to those the user has already interacted with,making it more scalable and effective for large datasets but struggling with new or unpopular 
items.User-based CF works well when user behavior is distinct,while Item-based CF is better when item relationships are strong.
```

```python
2)What is collaborative filtering and how does it work?
            Collaborative Filtering (CF) is a recommendation technique that suggests items based on user interactions.It works by finding similar
users(User-Based CF ) or similar items (item-based CF) to make recommendations without needing explicit item details.
```
