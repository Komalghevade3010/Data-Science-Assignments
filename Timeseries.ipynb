```python
from sklearn.metrics import mean_squared_error
```

```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
from sklearn.metrics import mean_squared_error



train = train.reset_index(drop=True)  
test = test.reset_index(drop=True)

train['t'] = np.arange(len(train))  
test['t'] = np.arange(len(train), len(train) + len(test))

# Fit linear regression model
linear_model = smf.ols('Ex_rate ~ t', data=train).fit()

# Predict on test data
pred_linear = linear_model.predict(test[['t']])

# Calculate RMSE
rmse_linear = np.sqrt(mean_squared_error(test['Ex_rate'], pred_linear))

# Print RMSE
print("RMSE:", rmse_linear)
```

```python

train['log_Ex_rate'] = np.log(train['Ex_rate'])

# Fit Exponential Model (OLS on log-transformed Ex_rate)
exp_model = smf.ols('log_Ex_rate ~ t', data=train).fit()

# Predict on test data
pred_log = exp_model.predict(test[['t']])
pred_exp = np.exp(pred_log)  # Convert predictions back to original scale

# Calculate RMSE
rmse_exp = np.sqrt(mean_squared_error(test['Ex_rate'], pred_exp))

print("RMSE (Exponential Model):", rmse_exp)
```

```python
train['t_square'] = train['t'] ** 2
test['t_square'] = test['t'] ** 2

# Fit Polynomial (Quadratic) Model
poly_model = smf.ols('Ex_rate ~ t + t_square', data=train).fit()

# Predict on test data
pred_poly = poly_model.predict(test[['t', 't_square']])

# Calculate RMSE
rmse_poly = np.sqrt(mean_squared_error(test['Ex_rate'], pred_poly))

# Print RMSE
print("RMSE (Polynomial Model):", rmse_poly)
```
