## Standardization of Numerical Variables ##

     Standardization is a scaling method where the values are centered around mean with a unit standard deviation.It means if we will calculate mean and standard deviation of standard scores it will be 0(mean) and 1(standard deviation) respectively.
     When we combine different magnitude columns in a single data set and when certain analysis is applied high magnitude column would dominate entire analysis.Low magnitude column will have a very little room to play analysis.so when you apply analysis on such dataset there would be a biasness in final analysis. so it remove this scaling effect,we will go for z-scores or data standardization.z-score range is -3 to 3.

```python
X=df['Volume']
X
```

```python
mean=df['Volume'].mean()
mean
```

```python
std_dev=df['Volume'].std()
std_dev
```

```python
z_Volume=(X-mean)/std_dev
z_Volume
```

## Conversion of Categorical Data into Dummy Variables ##

           If our dataset contains categorical data then there is no possible to apply machine learning algorithm because ML algorithm 
performs only on numerical columns.So we need to convert that categoricalmcolumns into the numerical ones.so in that case we use Dummy variable 
method on that columns and get the numerical columns.


```python
df=pd.get_dummies(data=df,columns=['Brand'])
```
