## Discuss the pros and cons of one hot encoding and label encoding ?##

```python
One hot encoding transforms categories into multiple binary columns,where each category is represented by o and 1.
This method is useful for nominal data ,ideal for tree based models like Random Forest and XGBoost.
On the other hand,Label Encoding,assigns a unique integer to each category,making it efficient in terms of storage and computation.
                                      It works well for ordinal data.
```
