```python
import numpy as np
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
df=pd.read_csv("C:/Users/gheva/OneDrive/Desktop/Neww Online Retail.csv",header=None)
df
```

```python
df.shape
```

```python
# Find unique values of column index 0.
items = (df[0].unique())
items
```

```python
encoded_vals = []
for index, row in df.iterrows(): # iterrows() returns index and content for each row
    labels = {}
    uncommons = list(set(items) - set(row))#total 9 items (Bagel,Milk,...,pencil) - items present in row = Bagel, Milk (for 1st row these 2 items are not present)
    commons = list(set(items).intersection(row))# common items
    for uc in uncommons:
        labels[uc] = 0 # assign 0 to uncommon items
    for com in commons:
        labels[com] = 1# assign 1 to common items
    encoded_vals.append(labels)
```

```python
encoded_vals
```

```python
ohe_df = pd.DataFrame(encoded_vals)
```

```python
ohe_df
```

```python
freq_items = apriori(ohe_df, min_support = 0.2, use_colnames = True)
```

```python
freq_items
```

```python
asso_rules=association_rules(freq_items, metric = "confidence", min_threshold = 0.6)
```

```python
asso_rules
```

```python
asso_rules.sort_values('confidence',ascending = False) 
```

## Interview Questions ##

# 1)What is lift and why it is important in association rules ?

```python
Ans:       Lift is a metric in association rule mining that measures how much more likely two items are to appear 
together compared to if they were independent.It is calculated as:
Lift=P(A intersection B)/(P(A)*P(B))
Lift>1: A and B occur together more than expected.
Lift=1: A and B are independent.
Lift<1: A and B occur together less than expected.

Importnace of lift in association rule:
.Helps identify strong,meaningful associations
.Filters out random co-occurrences
.Useful in market basket analysis to find valuable item combinations
```

#2)What is support and confidence .How do you calculate them?

```python
Ans:           Support: Support measures how frequently an itemset appears in the dataset.It is Calculated as:
Support= Transactions containing (A intersection B)/Total transactions

Confidence: Confidence measures the likelihood of item A appearing in a transaction given that item A is present.It is calculated as:

Confidence= Transactions containing(A intersection B)/Transactions containing A
```

# 3)What are some limitations or challenges of association rules mining?

```python
Ans: Limitations of Association rule mining:
1. Computational Complexity-Processing large datasets is slow.
2.Redundant Rules-Many rules may be trivial or irrelevant
3.Threshold Selection-Difficult to set optimal support and confidence.
4.Rare item Problem- Important but infrequent items may be ignored
5.Scalability Issues-Perfrmance drops with datas
```
